```python
import pandas as pd
var = pd.DataFrame({"Day":[1, 2, 3, 4, 5, 6],"Student_name":["rahul", "Sahil", "Prateek", "abhi", "vishali", "renu"], "Maths":[25, 15, 19, 43, 44, 49], "Dsa":[19, 34, 23, 28, 41, 42]})
var
```

# Melt

```python
pd.melt(var, id_vars=["Day"], var_name="Subjects", value_name="Marks") # This will make the df vertical and make the day col as the id/index and the variable name is set to subjects
```

# Pivot

```python
# var.pivot(index="Day", columns="Student_name") # This will set the day column as index and student name as column and maths and dsa as headers
var.pivot(index="Day", columns="Student_name", values="Dsa") # It will only give dsa values
```

```python
var.pivot_table(index="Student_name", columns="Day", aggfunc="mean") # This will give the mean 
```
