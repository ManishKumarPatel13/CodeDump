# Arithmetic Operations

```python
import pandas as pd
var = pd.DataFrame({"A":[1,2,3,4,5], "B":[6, 7, 8, 9, 10]})
```

```python
# Addition

var["sum"] = var["A"] + var["B"]
var
```

```python
# Subtraction
var["Sub"] = var["A"] - var["B"]
var
```

```python
# Multiplication
var["mul"] = var["A"] * var["B"]
var
```

```python
# Division
var["Div"] = var["A"] / var["B"]
var
```

```python
# Existence Checker
var["Checker"] = var["A"] <= 3 
var
```
