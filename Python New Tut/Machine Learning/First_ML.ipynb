```python
import pandas as pd
data = pd.read_csv("Solibility.csv")
data
```

```python
# Data Separation
X = data.drop("logS", axis=1)
Y = data["logS"]
```

```python
# Data Splitting 
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(X, Y, test_size=0.2, random_state=10)
train_x
```

```python
# Model Building 
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(train_x, train_y)
predict = model.predict(train_x)
predict
pre_test = model.predict(test_x)
```

```python
# Model Performance
from sklearn.metrics import mean_squared_error, r2_score
train_mean = mean_squared_error(train_y, predict)
r2_train = r2_score(train_y, predict)
r2_test = r2_score(test_y, pre_test)
test_mean = mean_squared_error(test_y, pre_test)
```

```python
# Printing Accuracy scores
# print(f"mean_squared_error (Train): {train_mean}")
# print(f"R2_Score (Train): {r2_train}")
# print(f"mean_squared_error (Test): {test_mean}")
# print(f"R2_Score (Test): {r2_test}")

# Making a Dataframe for these scores
Performance = pd.DataFrame(["Linear Regression", train_mean, r2_train, test_mean, r2_test]).transpose()
Performance.columns = ["Model", "MSB Train", "R2 Train", "MSB Test", "R2 Test"]
Performance
```
