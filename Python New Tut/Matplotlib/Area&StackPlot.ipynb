```python
import matplotlib.pyplot as plt
import numpy as np
```

# Stack Plot

```python
var1 = np.arange(1, 20)
var2 = np.random.randint(1, 20, 19)
var3 = np.random.randint(1, 20, 19)
var4 = np.random.randint(1, 20, 19)
var2
```

```python
plt.stackplot(var1, var2, var3, var4, labels=['f', 's', 't', 'fo'])
plt.legend()
plt.show()
```

```python
# Change color
plt.stackplot(var1, var2, var3, var4, labels=['f', 's', 't', 'fo'], colors=['r', 'g', 'b'])
plt.legend()
plt.show()
```

```python
# Change Baseline
plt.stackplot(var1, var2, var3, var4, labels=['f', 's', 't', 'fo'], baseline="sym") # make 0 the middle value
plt.legend()
plt.show()
```
