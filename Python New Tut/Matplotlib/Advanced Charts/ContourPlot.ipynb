```python
import matplotlib.pyplot as plt
from matplotlib import projections
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

# Contour Plots 
it is used to show a 3D surface on a 2D Plane

```python
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X,Y = np.meshgrid(xlist, ylist) # is used to create a rectangular grid of coordinates from two one-dimensional arrays xlist and ylist
Z = np.sqrt(X**2 + Y**2) # any function we want to plot our graph on
fig, ax = plt.subplots(1, 1)
c = ax.contour(X, Y, Z, 8, colors='black') # it is used to draw unfilled contour lines 
cp = ax.contourf(X, Y, Z, 8, cmap=plt.cm.hot) # {{x-co-ordinates}, {y-co_ordinates}, {height of plot or z co_ordinates}} all values in 2D format
ax.clabel(c, inline=1, fontsize=10) # it will label the values according to contour lines
fig.colorbar(cp)
plt.show()
```

```python
# Create a meshgrid of coordinates
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)

# Create a 2D array of temperature values
T = np.random.rand(100, 100)

# Plot the filled contour plot
plt.contourf(X, Y, T, cmap=plt.cm.hot)
plt.colorbar()
plt.title('Temperature Distribution')
plt.show()
```
