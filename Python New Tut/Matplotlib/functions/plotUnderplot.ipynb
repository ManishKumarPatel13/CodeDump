```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# How to Create a subplot within a subplot ?


```python
 # using gridspec
from matplotlib.gridspec import GridSpec

# Create a Figure object
fig = plt.figure()

# Create a GridSpec object
gs = GridSpec(2, 2)

# Create a subplot within the first row and first column of the grid
ax1 = fig.add_subplot(gs[0, 0])

# Create a subplot within the second row and first column of the grid
ax2 = fig.add_subplot(gs[1, 0])

# Create a subplot within the first row and second column of the grid
ax3 = fig.add_subplot(gs[0, 1])

# Create a subplot within the second row and second column of the grid
ax4 = fig.add_subplot(gs[1, 1])

# Plot some data in each subplot
ax1.plot([1, 2, 3], [4, 5, 6])
ax2.plot([7, 8, 9], [10, 11, 12])
ax3.plot([13, 14, 15], [16, 17, 18])
ax4.plot([19, 20, 21], [22, 23, 24])

# Set the axis labels for each subplot
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax4.set_xlabel('X')
ax4.set_ylabel('Y')

# Show the plot
plt.show()
```

```python
# using add_axes
x = np.arange(8)
s1 = np.random.randint(0, 8, 8)
s2 = np.random.randint(0, 8, 8)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1]) # {{[x-cord, y-co-ord, height, width]}}
inner_ax = fig.add_axes(rect=[0.7, 0.7, 0.25, 0.25])
ax.plot(x, s1)
inner_ax.plot(x, s2)
plt.show()
```
