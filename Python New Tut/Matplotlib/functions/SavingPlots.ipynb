```python
import matplotlib.pyplot as plt
import numpy as np
```

# Saving plots

```python
x = np.arange(5)
y = np.random.randint(0, 5, 5)
plt.plot(x, y)

plt.xticks(x, color='white')
plt.yticks(x, color='white')
# as png
# plt.savefig('line1.png', dpi=2000) # Providing dots per inch para. more dpi means clear graph
# plt.savefig('line1.png', facecolor='black') # facecolor changes the color of outside of the graph
plt.savefig('line1.png', facecolor='black', transparent=True) # Transparent fills the plot with the facecolor
plt.savefig('line2.png', facecolor='black', bbox_inches='tight') # Change the box inch


# as pdf
plt.savefig('line1pdf.pdf', dpi=2000, facecolor='black')

plt.show()
```
