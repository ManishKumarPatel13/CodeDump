```python
import matplotlib.pyplot as plt
import numpy as np
```

# Fill in between two plots

```python
x = np.arange(1, 20)
y = np.arange(1, 20 )
plt.plot(x, y, color='r')
# plt.fill_between(x=[4, 10], y1=4,y2=10, color='y') # Shows error if only one axis co-ordinates are given
plt.fill_between(x, y, color='y',where=(x>=2) & (x<=10))  # Using where keyword we can fill with using one co-ordinate also
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
