```python
import matplotlib.pyplot as plt
import numpy as np
```

# Working with texts

```python
x = np.arange(5)
y = np.array([3, 4, 2, 5, 1])

plt.plot(x,y)
plt.title('Normal Plot', fontsize=15)
plt.xlabel("days")
plt.ylabel('values')

# plt.text(0.9,4.1,'python', fontsize=15, style='italic') # {{x co-ord,y co-ord, 'text'}}
# plt.text(0.9,4.1,'python', style='italic', bbox={'facecolor':'red'}) # Creates text inside a box


plt.annotate('python', xy=(2,2), xytext=(1, 4.1), arrowprops=dict(facecolor='black', shrink=100)) # Creates an arrow from 1, 4.1 and pointing to 2,2 with black as facecolor

plt.legend(['up'], loc=2, edgecolor='black', shadow=True, framealpha=0.6) # Change edge color and all


plt.show()
```
