```python
import matplotlib.pyplot as plt
import numpy as np
```

# SubPlots

```python
x = np.arange(1, 10)
y = np.arange(1, 10)
plt.subplot(2, 2, 1) #  num must be an integer with 1 <= num <= 2, not 3
plt.plot(x, y)

plt.subplot(2, 2, 2) # {nrows, ncolums, index}
plt.pie([10, 20 ,30, 40])
plt.subplot(2, 2, 3)
plt.bar(['a', 'b', 'c', 'd'], [10,20,30,40])
plt.show()
```
