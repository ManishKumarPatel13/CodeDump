```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdate
import numpy as np
import datetime as dt
```

```python
months = mdate.MonthLocator() # it will extract month value from date
day = mdate.DayLocator() # it will extract year value from date
timefmt = mdate.DateFormatter(r"%Y-%m") # It will be the format of the data we want to extract
events = [dt.date(2023, 1, 23),dt.date(2023, 1, 28),dt.date(2023, 2, 3),dt.date(2023, 2, 21),dt.date(2023, 3, 15),
          dt.date(2023, 3, 24),dt.date(2023, 4, 8),dt.date(2023, 4, 24)]
readings = [12, 22, 14, 17, 19, 29, 16, 18] # Test readings
fig, ax = plt.subplots()
plt.plot(events, readings)
ax.xaxis.set_major_locator(months) # This will set the month as the main value in x axis
ax.xaxis.set_major_formatter(timefmt) # This will set how the values in x axis will look
ax.xaxis.set_minor_locator(day) # This will set the minor value in x axis
```
