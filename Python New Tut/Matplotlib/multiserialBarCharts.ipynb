```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Multiserial Bar Charts

```python
index = np.arange(5)
values1 = [5, 7, 9, 8, 3]
values2 = [4, 6, 8, 7, 2]
values3 = [3, 5, 7, 6, 1]
bw = 0.3
plt.axis([0,5,0,10])
plt.bar(index, values1, bw, color='b')
plt.bar(index+bw, values2, bw, color='g')
plt.bar(index+2*bw, values3, bw, color='r')
plt.xticks(index+1.5*bw, ['A', 'B', 'C', 'D', 'E'])
plt.savefig("multibar.png", dpi=3000)
```

# Multiserial bar charts with pandas DataFrame

```python
data = pd.DataFrame({'Series1':[1, 3, 4, 3, 5], 'Series2':[2, 4, 5, 2, 4], 'Series3':[3, 2, 3, 1, 3]})
data.plot(kind='bar').set_xticks(index,["A", "B", "C", "D", "E"])
```

# Stacked Bar Charts

```python
# Using pd
data.plot(kind='bar', stacked=True)
```

# Comparision Bar

```python
x0 = np.arange(8)
y1 = np.array([1, 3, 4, 6, 4, 3, 2, 1])
y2 = np.array([1, 2, 5, 4, 3, 3, 2, 1])
plt.ylim(-7, 7)
plt.bar(x0, y1, facecolor='r')
plt.bar(x0, -y2, facecolor='g') # One value of comparing factor should be negative
plt.xticks(())
plt.grid(True)

# Providing labels according to bar values
for x, y in zip(x0, y1):
    plt.text(x, y, '%d' % y, ha='center', va='bottom')
for x, y in zip(x0, y2):
    plt.text(x, -y, '%d' % y, ha='center', va='top')

plt.show()
```
