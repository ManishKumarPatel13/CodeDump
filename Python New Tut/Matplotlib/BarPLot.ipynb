# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['python', 'c', 'cpp', 'java', 'javascript']
y = [85, 70, 60 , 82, 84]
z = [90, 71, 80, 88, 89]

plt.bar(x, y)
plt.title("Programming Language Popularity", )
plt.xlabel("Programming Languages", fontsize=10)
plt.ylabel("Popularity -->")
plt.show()
```
