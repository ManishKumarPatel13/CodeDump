# Scatter Plot

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 3, 1, 4, 7]
y2 = [4, 6, 7, 2, 3]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("day")
plt.ylabel("NO.")
plt.show()
```
