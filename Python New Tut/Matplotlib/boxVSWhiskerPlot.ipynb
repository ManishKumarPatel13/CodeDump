```python
import matplotlib.pyplot as plt
import numpy as np
```

# Box

The centre line represents the median of the data set

# Whisky
The upper and lower part of a box plot which shows the highest and lowest value of a box plot

# outlier
When the common difference between two values is greater then the common difference between other values in the data set then a oval shape outside the whisky plot represents the plot

```python
var1 = np.arange(1, 20)
var1
```

```python
plt.boxplot(var1)
plt.show()
```

# Parameters

```python
# To Change the shape
# plt.boxplot(var1, notch=True)

# Change orientation
plt.boxplot(var1, vert=False)
plt.show()
```

```python
# Change Width
plt.boxplot(var1, widths=0.1)
plt.show()
```

```python
# Change the labels
plt.boxplot(var1, labels=["range"]) # Always provide as a list
plt.show()
```

```python
# Fill color inside the box
plt.boxplot(var1, patch_artist=True)
plt.show()
```

```python
# Show mean value
plt.boxplot(var1, showmeans=True) # Shown by a triangle in the middle of the median line
plt.show()
```

```python
# Show whisk
var2 = np.array([1, 2, 3, 4, 5, 12])
plt.boxplot(var2)
plt.show()
```

```python
# Join the line to whisk
plt.boxplot(var2, whis=3) # The last value to which the common difference becomes greater is muliplied with the whis if the value is equal or greater then the bigger value it joins the line 
plt.show()
```

```python
# Change outlier color
plt.boxplot(var2, sym='g')
plt.show()
```

```python
# Change box properties
plt.boxplot(var2, boxprops=dict(color='r'))
plt.show()
```

```python
# Change cap properties
plt.boxplot(var2, capprops=dict(color='r'))
plt.show()
```

```python
# Change whisk properties
plt.boxplot(var2, whiskerprops=dict(color='r'))
plt.show()
```

```python
# Change filler properties
plt.boxplot(var1,showmeans=True, flierprops=dict(markeredgecolor='r'))
plt.show()
```

```python
# Make multiple box plots
var = [var1, var2]
plt.boxplot(var, labels=["First", "second"])
plt.show()
```
