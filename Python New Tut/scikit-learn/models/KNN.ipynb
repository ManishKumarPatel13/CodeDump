```python
from sklearn.neighbors import KNeighborsClassifier 
import numpy as np
from sklearn import datasets
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

# KNN - K-Nearest-Neighbours

The K-nearest neighbors (KNN) classifier in scikit-learn is a simple, yet powerful machine learning algorithm for classification and regression tasks. It works by finding the K most similar examples in the training data to a new example and predicting the class or value of the new example based on the classes or values of the K most similar examples.

The KNN classifier is a non-parametric algorithm, which means that it does not make any assumptions about the distribution of the data. This makes it a good choice for problems where the data distribution is unknown or complex.

```python
iris = datasets.load_iris()
x = iris.data
y = iris.target # This is the target data we want to achieve
i = np.random.permutation(len(x)) # We are randomly changing the index positions so that we will have mixed data in our test set
x_train = x[i[:-10]] # Distributing values leaving 10 values for test and all other values is going in the test set
y_train = y[i[:-10]]
x_test = x[i[-10:]]
y_test = y[i[-10:]]
knn = KNeighborsClassifier() 
knn.fit(x_train, y_train) # Training the model with our dataset
pre = knn.predict(x_test)
print(accuracy_score(y_test, pre))
print(pre, y_test)



```

# Plot a graph

```python
sepal_lw = x[:,:2]
x_min, x_max = sepal_lw[:,0].min() - .5, sepal_lw[:,0].max() + .5
y_min, y_max = sepal_lw[:,0].min() - .5, sepal_lw[:,0].max() + .5
h = 0.2
cmap_light = ListedColormap(['r', 'b', 'g'])
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
knn2 = KNeighborsClassifier()
knn2.fit(sepal_lw, y)
Z = knn2.predict(np.c_[(xx.ravel(), yy.ravel())])
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

plt.scatter(sepal_lw[:,0], sepal_lw[:,1], c=y)
# plt.xlim(xx.min(), xx.max())
# plt.ylim(yy.min(), yy.max())

```
