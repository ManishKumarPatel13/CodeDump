```python
import sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25) # first two are the training sets in which model is trained and the other two are the sets in which it will be tested
clf = LogisticRegression() # it is a supervised learning algorithm to train data
clf.fit(X_train, y_train) # it is used to fit the data into the model
y_pred = clf.predict(X_test) # it predicts the new data on the model
accuracy = accuracy_score(y_test, y_pred) # Gets the accuracy of the model
print(accuracy)

```

```python
# scatter plot for sepal length and width of different species of iris
x = iris.data[:,0] # sepal length
y = iris.data[:,1] # sepal width

x_min, x_max = x.min() - .5, x.max() + .5
y_min, y_max = y.min() - 0.5, y.max() + 0.5
species = iris.target
plt.scatter(x, y, c=species)
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```
